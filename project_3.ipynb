OUTLIER DETECTION
(Interquartile Range (IQR) method )

```python
Q1 = df['product_price'].quantile(0.25)
Q3 = df['product_price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['product_price'] < lower) | (df['product_price'] > upper)]
print(outliers.shape)
```

```python
sns.boxplot(x=df['product_price'])
plt.show()
```

No outliers were detected in the dataset using the IQR method, as all values fall within the defined lower and upper bounds.

```python
Q1 = df['final_price'].quantile(0.25)
Q3 = df['final_price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['final_price'] < lower) | (df['final_price'] > upper)]
print(outliers.shape)
```

```python
sns.boxplot(x=df['final_price'])
plt.show()
```

The box plot of final_price reveals the presence of several high-value outliers beyond the upper whisker, indicating a subset of transactions with significantly higher purchase amounts compared to the majority.

Z-score method 

```python
mean = np.mean(df['product_price'])
std = np.std(df['product_price'])

z_scores = (df['product_price'] - mean) / std

outliers = df[(z_scores > 3) | (z_scores < -3)]
print(outliers.shape)
```

```python
mean = np.mean(df['final_price'])
std = np.std(df['final_price'])

z_scores = (df['final_price'] - mean) / std

outliers = df[(z_scores > 3) | (z_scores < -3)]
print(outliers.shape)
```

```python
sns.boxplot(x=df['final_price'])
plt.show()
```

The Z-score/IQR analysis identified 503 outliers in the product_price variable, indicating the presence of significantly high or low-priced products compared to the overall distribution.
